```python
# 经纬度 与 米 的换算

# GCS_WGS_1984
# 纬度：
# 1米≈1/(6378137*0.017453292519943299)°≈8.983152841195212e-06
# 1km = 0.008983152841195212  500m = 0.004491576420597606  250m = 0.002245788210298803  100m = 0.0008983152841195212
# 经度：
#   - 赤道：
#     - 1经度≈6378137*0.017453292519943299≈111319.4907932736米≈111.3194907932736千米
#     - 1米≈1/(6378137*0.017453292519943299)°≈8.983152841195212e-06°纬度
#   - 非赤道：
#     - 1经度≈(6378137*0.017453292519943299)*cosθ≈111319.4907932736*cosθ米≈111.3194907932736*cosθ千米（θ为纬度）
#     - 1米≈1/(111319.4907932736*cosθ)°经度

import math

def cos_angle(x):
    
    # 求sin()
    sin_mt = math.sin(math.radians(x))
    # 求cos()
    # 备注：直接用math.cos()出来的结果不完全正确
    cos_mt = math.sqrt(1-sin_mt**2)
    return cos_mt

width = 1000  # 单位m
height = 1000  # 单位m
latitude = 23.02   # 单位°
print(f"纬度: {latitude}°")
print(f"{width}米 = {width*1/(111319.4907932736*cos_angle(latitude))}°经度")
print(f"{width}米 = {width*1/(6378137*0.017453292519943299)}°纬度")
```
