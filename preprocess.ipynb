```python
import pandas as pd
import numpy as np
import geopandas as gpd

"""
车辆类型：
K1（大型客车），K2（中型客车），K3（小型客车），K4（微型客车）
H1（重型货车），H2（中型货车），H3（轻型货车），H4（微型货车）
C（公交车），D（出租车），qt（其他车型包含无法识别的车型）

燃料种类：
A（汽油），B（柴油），C（纯电），E（天然气），F（液化石油），O（混合动力），P（氢能），qt（其他）
"""
```

```python
# # 读取佛山市卡口设备信息
# bayonets = gpd.read_file(r'C:\01 毕业论文\7.论文代码和结果\硕士毕业论文代码\应用案例分析\佛山行政区和路网数据\中心城区路网和卡口\中心城区卡口点位.shp')
# 
# # 读取原始数据
# df = pd.read_csv(r'D:\24216373郑金涛中微观研究\2022-06-18_24卡口数据.csv')
# df['CLLX'] = df['CLLX'].str[:2]   # 车辆类型格式转换
# df['KKBH'] = df['KKBH'].astype('str')  # 卡口编号格式转换
# totalveh_number = len(df)
# print('所有车辆的原始数据量:', totalveh_number)
# 
# df = pd.merge(df, bayonets[['KKBH']], on='KKBH', how='inner')
# print('中心城区的卡口数据:', len(df), '匹配成功率: {:.2%}'.format(len(df) / totalveh_number))
# 读取卡口点位
bayonets = gpd.read_file(r'C:\01 毕业论文\7.论文代码和结果\硕士毕业论文代码\应用案例分析\佛山行政区和路网数据\中心城区路网和卡口\中心城区卡口点位.shp')
bayonets['KKBH'] = bayonets['KKBH'].astype(str).str.strip()
bayonet_ids = set(bayonets['KKBH'])

# 初始化
filtered_chunks = []
chunksize = 500_000
totalveh_number = 0
matched_number = 0

# 分块读取
csv_path = r'D:\24216373郑金涛中微观研究\2022-06-18_24卡口数据.csv'
for i, chunk in enumerate(pd.read_csv(csv_path, chunksize=chunksize)):
    # 去除空值
    chunk = chunk.dropna(subset=['GCRQ', 'GCSJ', 'KKBH', 'FZJG', 'CLTMBH']).copy()
    chunk['CLLX'] = chunk['CLLX'].astype(str).str[:2]
    chunk['KKBH'] = chunk['KKBH'].astype(str).str.strip()
    totalveh_number += len(chunk)

    chunk_filtered = chunk[chunk['KKBH'].isin(bayonet_ids)]
    matched_number += len(chunk_filtered)
    filtered_chunks.append(chunk_filtered)

    print(f"第 {i+1} 块处理完成，当前累计匹配数: {matched_number}")

# 合并所有过滤结果
df = pd.concat(filtered_chunks, ignore_index=True)

# 输出统计信息
print('所有车辆的原始数据量:', totalveh_number)
print('中心城区的卡口数据:', len(df), '匹配成功率: {:.2%}'.format(len(df) / totalveh_number))
```

```python
# 筛选研究车辆
valid_cllx = ['K1', 'K2', 'K3', 'K4', 'H1', 'H2', 'H3', 'H4']  # 客车、货车
valid_rllx = ['A', 'B']   # 汽油、柴油
valid_df = df[df['CLLX'].isin(valid_cllx)]
valid_df = valid_df[valid_df['RLLX'].isin(valid_rllx)]
valid_df = valid_df.reset_index(drop=True)
origin_volume = len(valid_df)
print('研究车辆的原始数据:', origin_volume, 
      '占比: {:.2%}'.format(origin_volume / len(df)))

# 客车
cars_df = valid_df[valid_df['CLLX'].isin(['K1', 'K2', 'K3', 'K4'])]
cars_petrol_df = cars_df[cars_df['RLLX'] == 'A']
cars_diesel_df = cars_df[cars_df['RLLX'] == 'B']
print('客车数量:', len(cars_df),
      '客车占比: {:.2%}'.format(len(cars_df) / origin_volume), 
      '汽油: {:.2%}'.format(len(cars_petrol_df) / len(cars_df)), 
      '柴油: {:.2%}'.format(len(cars_diesel_df) / len(cars_df)))

# 货车
trucks_df = valid_df[valid_df['CLLX'].isin(['H1', 'H2', 'H3', 'H4'])]
trucks_petrol_df = trucks_df[trucks_df['RLLX'] == 'A']
trucks_diesel_df = trucks_df[trucks_df['RLLX'] == 'B']
print('货车数量:', len(trucks_df),
      '货车占比: {:.2%}'.format(len(trucks_df) / origin_volume), 
      '汽油: {:.2%}'.format(len(trucks_petrol_df) / len(trucks_df)), 
      '柴油: {:.2%}'.format(len(trucks_diesel_df) / len(trucks_df)))

valid_df.head(3)
```

```python
# 剔除车辆信息缺失的数据
valid_df = valid_df.dropna(subset=['CLLX'])
print('剔除车辆信息缺失后的数据量: ', len(valid_df), round(len(valid_df) / origin_volume, 4))

# 提取佛山市本地车牌
valid_df = valid_df[(valid_df['FZJG'] == '粤E') | (valid_df['FZJG'] == '粤X') | (valid_df['FZJG'] == '粤Y')]
print('佛山市本地车辆的数据量: ', len(valid_df), round(len(valid_df) / origin_volume, 4))

# 数据格式处理
valid_df['GCRQ'] = valid_df['GCRQ'].str.split(' ', expand=True)[0]   # 日期格式转换
valid_df['GCSJ'] = pd.to_datetime(valid_df["GCSJ"])     # 时间格式转换
valid_df = valid_df.drop_duplicates(subset=['GCRQ', 'GCSJ', 'CLTMBH'])    # 删除重复过车数据
valid_df = valid_df.sort_values(by=['GCRQ', 'GCSJ', 'CLTMBH'])    # 根据车牌号、出行日期、出行时间排序
valid_df = valid_df.reset_index(drop=True)                       # 重建索引
print('数据格式转换后的数据量: ', len(valid_df), round(len(valid_df) / origin_volume, 4))

valid_df.head(3)
```

```python
# 统计能够与路网实际卡口和拓扑节点匹配的卡口记录数

# valid_df['KKBH'] = valid_df['KKBH'].astype('str')

# 读取与路网节点匹配成功的卡口设备
pos_bayonets = pd.read_csv(r'C:\01 毕业论文\7.论文代码和结果\硕士毕业论文代码\应用案例分析\车辆路径重构\1.构建路网拓扑结构\center_realpos.csv')
pos_bayonets['KKBH'] = pos_bayonets['KKBH'].astype('str')

# 计算匹配率
pos_mapping_df = pd.merge(valid_df, pos_bayonets[['KKBH']], on='KKBH', how='inner')
pos_mapping_df = pd.merge(pos_mapping_df, bayonets[['KKBH', 'Longitude', 'Latitude']], on='KKBH', how='left')  # 匹配卡口的经纬度

print('与真实路网卡口和路网节点均匹配成功的数据:', len(pos_mapping_df), '匹配成功率: {:.2%}'.format(len(pos_mapping_df) / len(valid_df)))
```

```python
# 按照车辆号牌、过车日期、过车时间排序
pos_mapping_df = pos_mapping_df.sort_values(by=['CLTMBH','GCSJ','GCRQ'  ])
pos_mapping_df.reset_index(drop=True, inplace=True)
pos_mapping_df.head(3)
```

```python
# 保存预处理后的数据
pos_mapping_df.to_csv(r'D:\24216373郑金涛中微观研究\center_valid_df.csv', index=False, encoding='gbk')
```

```python
# 保存车辆信息表
vehicle_info = pos_mapping_df.drop_duplicates(subset='CLTMBH', keep='first')
vehicle_info = vehicle_info[['CLTMBH', 'HPZL', 'CLLX', 'RLLX', 'PFBZ', 'SYXZ']]   #pfbz、syxz改为大写

# 创建CLLX、RLLX与中文车辆类型、燃料类型的对照表
cllx_mapping = {'K1': '大型客车',
                'K2': '中型客车',
                'K3': '小型客车',
                'K4': '微型客车',
                'H1': '重型货车',
                'H2': '中型货车',
                'H3': '轻型货车',
                'H4': '微型货车'}
rllx_mapping = {'A': '汽油', 'B': '柴油'}

# 中文映射
vehicle_info['veh_type'] = vehicle_info['CLLX'].map(cllx_mapping)
vehicle_info['fuel_type'] = vehicle_info['RLLX'].map(rllx_mapping)

print('研究车辆数:', len(vehicle_info))
vehicle_info.to_csv(r'D:\24216373郑金涛中微观研究\中心城区研究车辆信息表.csv', index=False, encoding='gbk')
vehicle_info.head(3)
```
